# Python List Comprehension


```
#Example 1: Print all the items in a list
mylist = ["iPhone", "Android", "Huawei", "Nokia", "HPC"]

#Foor Loop
for x in mylist:
  print("Phone Brand:", x)
```

```
range(len(mylist))
```

```
#Index Numbers on List to access values
mylist = ["iPhone", "Android", "Huawei", "Nokia", "HPC"]

for x in range(len(mylist)):
  print("Phone Brand:", mylist[x])
```

```
#Extract list values using a while loop
mylist = ["iPhone", "Android", "Huawei", "Nokia", "HPC"]

x = 0

while x < len(mylist):
  print("Phone Brand:", mylist[x])
  x+=1
```

```
#List Comprehension
mylist = ["Imposters Among Us", "COD", "Don't Starve Together", "Plants vs. Zombies"]

[print(x) for x in mylist]
```

```
#Find all the games with the letter m contained inside, and show them to me
mylist = ["Imposters Among Us", "COD", "Don't Starve Together", "Plants vs. Zombies"]

newlist = []

for x in mylist:
  if 'm' in x:
    newlist.append(x)

print(newlist)
```

```
#With List Comprehension
mylist = ["Imposters Among Us", "COD", "Don't Starve Together", "Plants vs. Zombies"]

newlst = [x for x in mylist if 'm' in x]
print(newlist)
```

```
#Iterable Lists with Lists Comprehension
#Using the range function to create int lists

newlist = [x for x in range(20)]
print(newlist)
```

```
newlist = [x for x in range(20) if x < 12]
print(newlist)
```

```
#Example, Convert everything to UPPERCASE
mylist = ["Books", "keys", "Wallet", "Laptop", "BooKbag", "GymBag"]

print(mylist)

newlist = [x.upper() for x in mylist]

print(newlist)
```

```
#Example: Set all the values to "Yep"

mylist = ["Books", "keys", "Wallet", "Laptop", "BooKbag", "GymBag"]

print(mylist)

newlist = ["Yep" for x in mylist]

print(newlist)
```

```
#If the value is bookbag, say KnapSack
mylist = ["Books", "keys", "Wallet", "Laptop", "BooKbag", "GymBag"]

print(mylist)

newlist = [x if x.lower() != "bookbag" else "KnapSack" for x in mylist]

print(newlist)
```

```
#Sorting Examples
#Example 1
mylist = ["Books", "Pens", "Pencils", "Bookbag", "Altoids", "Camera", "Deodorant"]

print(mylist)

mylist.sort()

print(mylist)
```

```
#Example 2
mylist = [23, 12, 94, 58, 27, 123, 5, 78, 29]

print(mylist)

mylist.sort()

print(mylist)
```

```
#Example 3
mylist1 = ["Books", "Pens", "Pencils", "Bookbag", "Altoids", "Camera", "Deodorant"]
mylist2 = [23, 12, 94, 58, 27, 123, 5, 78, 29]
mylist3 = [100, 29, 449, 20, 584, 57]

print(mylist1)
mylist1.sort(reverse = True)
print(mylist1)

print(mylist2)
mylist2.sort(reverse = True)
print(mylist2)

print(mylist3)
mylist3.sort(reverse = True)
print(mylist3)
```

```
#Case Sensitive Sorting with Lists
mylist = ["Apples", "bananas", "Oranges", "kiwis", "Cherries", "peaches"]

print(mylist)

mylist.sort()

print(mylist)
```

```
mylist = ["Apples", "bananas", "Oranges", "kiwis", "Cherries", "peaches"]

print(mylist)

mylist.sort(key = str.lower)

print(mylist)
```

```
#Reverse method
mylist = ["Sandwiches", "Heroes", "Gyros", "Melts", "Submarines"]

print(mylist)
mylist.reverse()
print(mylist)
```

```
#Copying Lists

mylist1 = ["Red", "White", "Blue"]
mylist2 = mylist1
mylist3 = list(mylist1)   #Forces Python to store memory in a different area
mylist4 = mylist1.copy()  #Forces Python to store memory in a different area
```

```
print(mylist1)
print(mylist2)
print(mylist3)
print(mylist4)
```

```
mylist2.pop() #Because mylist1 and mylist2 is in the same memory location both changes
print(mylist1)
print(mylist2)
print(mylist3)
print(mylist4)
```

```
mylist3.pop(0)
print(mylist1)
print(mylist2)
print(mylist3)
print(mylist4)
```

```
#Example Appending two lists
mylist1 = ["One", "two", "three"]
mylist2 = [1,2,3]

for x in mylist2:
  mylist1.append(x)

print(mylist1)
```
