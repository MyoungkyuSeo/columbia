```
#While Loop 
#Print i as long as i is less than 10 

i = 1

while i < 10:
  print("The loop iteration number", i)
  #must increment the iterator or we will have infinite loop
  i += 1
```

```
#While loop with breaks

i = 1

while i < 20:
  if i == 13:
    print("We have reached UNLUCKY number", i, "WE MUST EXIT NOW!")
    break
  
  print("The loop iteration number", i)

  #must increment the iterator or we will have infinite loop
  i += 1
```

```
#While loop with continue
i = 1

while i < 20:

  if i == 10:
    print("We have reached UNLUCKY number", i, "WE MUST EXIT NOW!")
    i+=1
    continue

  if i == 13:
    print("We have reached UNLUCKY number", i, "WE MUST EXIT NOW!")
    break
  
  print("The loop iteration number", i)

  #must increment the iterator or we will have infinite loop
  i += 1
```

```
#While loop with else
i = 1

while i < 20:

  if i == 10:
    #print("We have reached UNLUCKY number", i, "WE MUST EXIT NOW!")
    i+=1
    continue

  if i == 13:
    print("We have reached UNLUCKY number", i, "WE MUST EXIT NOW!")
    #break
  
  print("The loop iteration number", i)

  #must increment the iterator or we will have infinite loop
  i += 1
else:
  print("While loop has completed. Thanks for counting with me.\
          I am not designed to count beyond 20")
```

```
#While Loop Example Program
#Tell me a Secret with authentication

secret = "A unicorn over the rainbow makes the morning crisper and the air sweeter"
```

```
#Login Attemps counter
i = 0
maxloginattempts = 10
#While loop conditional
condition = True


while condition:
  print("Hello, I have a secret to tell you, but you must login first: You have", maxloginattempts-i, "Login Attempts")

  if i == maxloginattempts:
    print("Too Many Login Attempts, I can't tell you now!")
    break

  #Ask for user name and password
  username = input("What is your Username: ")
  password = input("What is your Password: ")

  if username == "RAZZAK" and password == "TellMe123":
    print("Shhhhhh, the secret is:", secret)
    #Turn off the condition, switch to false
    condition = False
  else:
    #Increase incorrect login attempt
    i+=1
    print("I don't regonize you! Try again!", "Login Attempt: ", i)
    continue
else:
  print("Thanks for Playing the Secret Game with me")
```

```
#For Loop
#Print Each Item in the list
groceries = ["apples", "bananas", "cherries", "avocados", "noodles", "pasta sauce"]
```

```
for x in groceries:
  print("Don't forget to buy:", x)
```

```
#For loop over a raw string

for x in "Madagascar":
  print("A letter in the string is:", x, "Then the next letter is below")
```

```
#Break out of for loop

for x in groceries:
  if x == "noodles":
    break

  print("Don't forget to buy:", x)
```

```
#Continue for loop

groceries = ["apples", "bananas", "cherries", "avocados", "noodles", "pasta sauce"]

for x in groceries:
  if x == "noodles":
    print("Spicy Korean noodles please, otherwise don't buy it")
    continue
  print("Don't forget to buy:", x)
```

```
#Range with starting params
for x in range(4, 8):
  print(x)
```

```
#range with starting, ending and by params
for x in range(0, 100, 5):
  print(x)
```

```
#range with starting, ending and by parans with else
for x in range(0, 120, 20):
  print(x)
else:
  print("Done printing 0 to 100 by 20")
```

```
#Nested Loops

people = ["John", "Jane", "January", "Jon", "Jonathan"]
adj = ["Great", "Sweet", "Fantastic", "Terrific", "Stupendous", "WOW!"]

for x in people: #outer loop
  #print(x)
  for y in adj: #inner loop / nested loop
    print(x, "is thought to be....just", y, "!!!!!")
```
