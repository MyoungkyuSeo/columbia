Python Lists


```
#Creating a List 

mylist = ["tomato", "potato", "my big toe"]

print(mylist, type(mylist))
```

```
#Allows for Duplicates
mylist = ["fish", "chicken", "beef", "pork", "plant-based stuff", 'fish', "fish"]
print(mylist)
print(type(mylist))
```

```
#Using the Len fuction on the list

mylist = ["Once", "Twice", "Thrice", "Fourice", "Fivice", "Sice"]

print(len(mylist))
```

```
#Example of Different Datatypes in Lists

list1 = ["bubble gum", "miint gum", "fresh gum"]
list2 = [1, 3, 45, 670, 3]
list3 = [True, False, True, True, False, True]

print(type(list1), type(list2), type(list3))
```

```
#Example of Mixed Datatypes in list
list1 = ["Alfred", 456, "Alf", True, 24.67]

print(type(list1))
```

```
mytuple = ("candy", 'fruits', "vegetables")
print(type(mytuple), mytuple)

mytuple = list(mytuple)

print(type(mytuple), mytuple)
```

```
#Acessing List Items
mylist = ["Toothpicks", "Toothpaste", 'Mouthwash']

print(mylist[1])
```

```
#Negative Index
mylist = ["Face Wash", "Toner", "Cleanser", 'Moisturizer']

print(mylist[-1])
print(mylist[-2])
print(mylist[-3])
print(mylist[-4])
#print(mylist[-5]) this gives error
```

```
#Range of Index
#Example of start and end range
mylist = ["SUV", "Sedan", "Compact", "Coupe", "Exotic", "Sports", "EV"]
```

```
#"Compact", "Coupe", "Exotic"
print(mylist[3:6])
print(mylist[2:5])
```

```
#Just put Ending position
print(mylist[0:5])
print(mylist[:5])
```

```
#Put beginning position
print(mylist[4:])
```

```
#Change items in a List
mylist = ["Drama", "Thriller", "Horror", "Documentary", "Comedy"]

print(mylist)

print(mylist[2])

mylist[2] = "RomCom"

print(mylist[2])

print(mylist)
```

```
#Creating nested lists, multidimensionl

mylist = ["Animated", "Live-Action", "CGI", "Hybrid"]

print(mylist)

mylist[1] = ['Live-Action:Motion-Captured', 'Live-Action:Special Effects']

print(mylist)
```

```
#print(mylist[1[1]])

print(mylist[1][1])
```

```
#Change a Range of Items in a List
mylist = ["Jetski", "ParaSailing", "Wakeboarding", "Surfing", "Snorkeling"]
print(mylist)

mylist[1:3]
mylist[1:3] = ["Beach Volleyball", "Sand Castles"]
print(mylist)
```

```
#Insert items into list
mylist = ["Hot Oatmeal", "Hot Cookies", "Hot Pie", 'Hot Soup']

print(mylist)

mylist.insert(2, "Hot Coco")

print(mylist)

mylist.insert(4, "Hot Fries")

print(mylist)
```

```
#Adding items to the end of list
print(mylist)

mylist.append("Hot Custard")

print(mylist)
```

```
#Insert in position item
print(mylist)

mylist.insert(1, "Hot Noodles")

print(mylist)
```

```
print(mylist)
warmlist = ["Warm Tomato Soup", "Warm Baked Potato", "Warm Applesauce", 'Warm Cheese']

print(warmlist)

mylist.extend(warmlist)

print(mylist)
```

```
#Extend a list with tuple
print(mylist)

mytuple = ("Apples", "Oranges")

print(type(mytuple), type(mylist))

mylist.extend(mytuple)

print(mylist)
```

```
#Removing Items from a List
#Remove method
print(mylist)

mylist.remove("Warm Tomato Soup")

print(mylist)
```

```
#pop() method
print(mylist)

mylist.pop(1)

print(mylist)
```

```
#pop() method
print(mylist)

mylist.pop()

print(mylist)
```

```
#Exampel of removing first item in list

print(mylist)

del (mylist[0])

print(mylist)
```

```
#Clear method

print(mylist)

mylist.clear()

print(mylist)
```

```
#Remove object using del
print(mylist)

del mylist

#print(mylist)
```

```
#While Loop 
#Print i as long as i is less than 10 

i = 1

while i < 10:
  print("The iteration number is:", i)
  i+=1
```
