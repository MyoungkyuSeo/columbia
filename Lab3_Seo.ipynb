```
#Comparison Operators
x = "Johnny Dough"
y = "John Doe"
z = "Johnny Doe"
i = 2456
j = 4677
k = 2456
m = -2456
```

```
#Equals = 
#Strings
print("Does", x, "have the same value as", y, "?:", x==y) #FALSE
print("Does", x, "have the same value as", z, "?:", x==z) #FALSE
print("Does", x, "have the same value as", x, "?:", x==x) #TRUE
print("Does", x, "have the same value as", "Johnny Dough", "?:", x== "Johnny Dough") #TRUE

#Numbers
print("Does", x, "have the same value as", i, "?:", x==i) #FALSE
print("Does", i, "have the same value as", k, "?:", i==k) #TRUE
print("Does", i, "have the same value as", m, "?:", i==m) #FALSE
```

```
#NOT Equals = 
#Strings
print("Does", x, "does not have the same value as", y, "?:", x != y) #TRUE
print("Does", x, "does not have the same value as", z, "?:", x != z) #TRUE
print("Does", x, "does not have the same value as", x, "?:", x !=x ) #FALSE
print("Does", x, "does not have the same value as", "Johnny Dough", "?:", x != "Johnny Dough") #FALSE

#Numbers
print("Does", x, "does not have the same value as", i, "?:", x!=i) #TRUE
print("Does", i, "does not have the same value as", k, "?:", i!=k) #FALSE
print("Does", i, "does not have the same value as", m, "?:", i!=m) #TRUE
```

```
#Greater than = 
#Strings
#Based on alphabetical order
print("Does", x, "is greater than", y, "?:", x > y)
print("Does", "F", "is greater than", "L", "?:", "F" > "L")
```

```
#Numbers
print("Is", i, "is greater than", i, "?:", i > i) #FALSE
print("Is", i, "is greater than", j, "?:", i > j) #FALSE
print("Is", i, "is greater than", k, "?:", i > k) #FALSE
print("Is", i, "is greater than", m, "?:", i > m) #true
```

```
#Less than
#Numbers
print("Is", i, "is less than", i, "?:", i < i) #FALSE
print("Is", i, "is less than", j, "?:", i < j) #TRUE
print("Is", i, "is less than", k, "?:", i < k) #FALSE
print("Is", i, "is less than", m, "?:", i < m) #FALSE
```

```
#Greater Than EQ
#Numbers
print("Is", i, "is greater than eq", i, "?:", i >= i) #TRUE
print("Is", i, "is greater than eq", j, "?:", i >= j) #FALSE
print("Is", i, "is greater than eq", k, "?:", i >= k) #TRUE
print("Is", i, "is greater than eq", m, "?:", i >= m) #TRUE

#Less Than EQ
#Numbers
print("Is", i, "is less than eq", i, "?:", i <= i) #TRUE
print("Is", i, "is less than eq", j, "?:", i <= j) #TRUE
print("Is", i, "is less than eq", k, "?:", i <= k) #TRUE
print("Is", i, "is less than eq", m, "?:", i <= m) #FALSE
```
