```
#Login Attemps counter

def mysecretgame(secret = "A unicorn over the rainbow makes the morning crisper and the air sweeter", username = "SEO", password = "TEST123", maxloginattempts = 10):
  i = 0
  #While loop conditional
  condition = True
  while condition:
    print("Hello, I have a secret to tell you, but you must login first: You have", maxloginattempts-i, "Login Attempts")

    if i == maxloginattempts:
      print("Too Many Login Attempts, I can't tell you now!")
      break

    #Ask for user name and password
    username_input = input("What is your Username: ")
    password_input = input("What is your Password: ")

    if username_input == username and password_input == password:
      print("Shhhhhh, the secret is:", secret)
      #Turn off the condition, switch to false
      condition = False
    else:
      #Increase incorrect login attempt
      i+=1
      print("I don't regonize you! Try again!", "Login Attempt: ", i)
      continue
  else:
    print("Thanks for Playing the Secret Game with me")
```

```
def LettertoNumberGradeRange(y = ""):
  if y == "":
    y = input("Input letter Grade")
  if not y.isdigit():
    if y == "D":
      print("Letter Grade D, 65-66")
    elif y == "D+":
      print("Letter Grade D+, 67-69")
    elif y == "C-":
      print("Letter Grade C-, 70-72")
    elif y == "C":
      print("Letter Grade C, 73-76")
    elif y == "C+":
      print("Letter Grade C+, 77-79")
    elif y == "B-":
      print("Letter Grade B-, 80 - 82")
    elif y == "B":
      print("Letter Grade B, 83-86")
    elif y == "B+":
      print("Letter Grade B+, 87-89")
    elif y == "A-":
      print("Letter Grade A-, 90-92")
    elif y == "A":
      print("Letter Grade A, 93-96")
    elif y == "A+":
      print("Letter Grade A+, 97-100")
    else:
      print("Please input grades from A+ to D")
  if y.isdigit():
    y = int(y)
    if y <= 66:
      print(y, "is Letter grade D")
    elif y >= 67 and y <= 69:
      print(y, "is Letter grade D+")
    elif y >= 70 and y <= 72:
      print(y, "is Letter grade C-")
    elif y >= 73 and y <= 76:
      print(y, "is Letter grade C")
    elif y >= 77 and y <= 79:
      print(y, "is Letter grade C+")
    elif y >= 80 and y <= 82:
      print(y, "is Letter grade B-")
    elif y >= 83 and y <= 86:
      print(y, "is Letter grade B")
    elif y >= 87 and y <= 89:
      print(y, "is Letter grade B+")
    elif y >= 90 and y <= 92:
      print(y, "is Letter grade A-")
    elif y >= 93 and y <= 96:
      print(y, "is Letter grade A")
    elif y >= 97 and y <= 100:
      print(y, "is Letter grade A+")
    else:
      print("Please input a number between 0 and 100")
```

```
#Part 2 of the Challenge
#Taxes Function

def simpletaxreturnfunction(income = 100000, tax = 24, deduction = 4000):
  taxes_owed = income * (tax/100)
  taxes_deducted = taxes_owed - deduction

  if taxes_deducted >= 0:
    print("You owe ", taxes_deducted, "in taxes")
  else:
    print("You will get a return of", -1 * taxes_deducted)

  return(taxes_deducted)
```

```
#Math function to calculate values of a triangle
import math
import cmath

def trig(sideA, angleA, sideB):
  print("Please type the angles in radians, and the answers in radians")
  B = (cmath.sin(angleA)/sideA)*sideB
  angleB = cmath.asin(B)
  angleC = math.pi-angleB-angleA

  sideC = sideA/cmath.sin(angleA)*cmath.sin(angleC)

  answer = {
            "sideA" : sideA,
            "angleA" : angleA,
            "sideB" : sideB,
            "angleB" : angleB,
            "sideC" : sideC,
            "angleC" : angleC,
                  }     
  print(answer)
```

```
trig(10, 0.610865, 11)
```

```
def trig2(sideA, angleA, sideB):
  print("Please input the angle as degrees, answer would be outputted as degrees")
  radA = math.radians(angleA)

  B = (math.sin(radA)/sideA)*sideB
  radB = math.asin(B)
  angleB = math.degrees(radB)

  angleC = 180-angleA-angleB
  radC = math.radians(angleC)

  sideC = sideA/math.sin(radA)*math.sin(radC)

  answer = {
            "sideA" : sideA,
            "angleA" : angleA,
            "sideB" : sideB,
            "angleB" : angleB,
            "sideC" : sideC,
            "angleC" : angleC,
                  }     
  print(answer)
```
