```
#4 With For Loops

mylist = ["Watermenlon", "Programming", 203040, "Exciting", "Astounding", "Teriffic"]
newlist = []

for i in mylist:
  if type(i) == int:
    newlist.append(i)
    continue
  list2 = "".join([x for x in i if str(x) not in 'aeiouAEIOU'])
  newlist.append(list2)
print(newlist)
```

```
#4 With For Loops(Simplified)
mylist = ["Watermenlon", "Programming", 203040, "Exciting", "Astounding", "Teriffic"]
newlist = []

for i in mylist:
  if type(i) == int:
    newlist.append(i)
    continue
  newlist.append("".join([x for x in i if str(x) not in 'aeiouAEIOU']))
print(newlist)
```
