```
#Data Types in Dictionaries 

datatype_dictionary = {
                        "FavoriteFruit" : "Tomato", #string
                        "Vegan" : False, #boolean
                        "Daily Calories" : 2500, #int
                        "FavoriteCarbs" : {"Penne", "Spagetti", "Sandwiches", "Hot Bread", "Noodles"}, #Tuple
                        "FavoriteProteins" : ["Steak", "Chicken", "Beans", "Chickpeas", "Plant-Based Poweder"] #List        

                          }
```

```
print("my favorite fruit is", datatype_dictionary["FavoriteFruit"])
print("Am I Vegan?", datatype_dictionary["Vegan"])
print("My caloric intake per day is", datatype_dictionary["Daily Calories"])
print("My favorite Carbs are", datatype_dictionary["FavoriteCarbs"])
```

```
print(type(datatype_dictionary["FavoriteFruit"]))
print(type(datatype_dictionary["Vegan"]))
print(type(datatype_dictionary["Daily Calories"]))
print(type(datatype_dictionary["FavoriteCarbs"]))
print(type(datatype_dictionary["FavoriteProteins"]))
```

```
#Using the get method on a pythond ictionary

x = datatype_dictionary.get("Daily Calories")
print(x)
print(datatype_dictionary["Daily Calories"])
```

```
#Get the key and values
keys = datatype_dictionary.keys()
print(keys)
```

```
values = datatype_dictionary.values() #Before any changes
print(values)
```

```
print(values)
datatype_dictionary["FavoriteFruit"] = {"Apples", "Kiwis", "Tomato"}
print(values)
```

```
#Using the Items method

x = datatype_dictionary.items()
print(x)

#Update the value for Calories
datatype_dictionary["Daily Calories"] = 3500

print(x)
```

```
#Using if then, and in statements determine if certain keys exist in a dictionary

if "model" in datatype_dictionary.keys():
  print(" 'model' key does exist")
elif "Favorite" in datatype_dictionary.keys():
  print("'Favorite' keys do exist")
elif "FavoriteFruit" in datatype_dictionary.keys():
  print("'FavoriteFruit' keys do exist")
else:
  print("The keys were not found")
```

```
#Copying Dictionaries 

copy_dict = {
              "Product" : "Jalapeo Chips",
              "SKU" : 86654239,
              "Cost" : 5.99,
              "Size" : "16oz",
              "Inventory" : True
              }
```

```
my_dict1 = copy_dict
my_dict2 = copy_dict.copy()
my_dict3 = dict(copy_dict)
```

```
#Edit copy_dict
print(copy_dict)
print(my_dict1)
print(my_dict2)
print(my_dict3)

copy_dict["Inventory"] = False
copy_dict["Product"] = "Sriracha Peas"
```

```
print(copy_dict) #Original, will have changes
print(my_dict1)  #Yes changes
print(my_dict2)  # No changes, using .copy
print(my_dict3)  # No changes, constructor function
```
