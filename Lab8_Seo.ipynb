```
#Arbritary Arguments

def my_fav_pet(*name):
  if len(name) >= 3:
   print("My third favorite pet name is", name[2])
  else:
    print("My least favorite name is", name[len(name)-1])
```

```
my_fav_pet("Tommy", "Mykonos", "Santorini", "Jellybeans", "Spot")
```

```
my_fav_pet("Tommy", "Mykonos")
```

```
#Keyword Arguments

def my_eat_drink_places(food, drink, place):
  print("My favorite food to eat is the", food, "because it makes me happy!")
  print("I enjoy drinking a", drink, "with my", food, "because it is refreshing and complimentary")
  print("But I especially enjoy eating my", food, "and drinking my", drink, "at the", place, "becuase its chill vibes")
```

```
#Example
my_eat_drink_places("Burger", "Milkshake", "Beach")
```

```
#Example 2
my_eat_drink_places(place = "Beach", food = "Burger", drink = "Milkshake")
```

```
#Arbritary Keyword Arguments

def my_fav_sports(**sports):
  if "favplayer" in sports:
    print("My favorite sports player is", sports["favplayer"])

  if "favsportwatch" in sports:
    print("My favorite sports to watch is", sports["favsportwatch"])

  if "favsportplay" in sports:
    print("My favorite sports to play is", sports["favsportplay"])

  print("This is the end of the function")
```

```
my_fav_sports(favplayer = "Messi", favsportwatch = "American Football", favsportplay = "Ultimate Freesbe", favsporttime = "Primetime", favsportplace = "Miami")
```

```
my_fav_sports(favplayer = "Messi", favsportwatch = "American Football", favsporttime = "Primetime", favsportplace = "Miami")
```

```
#Default Parameters
def my_fav_vacation(country = "Jamaica", season = "Spring"):
  print("The country I enjoy vacationing at the most is", country, "during the", season)
```

```
#Call the function
my_fav_vacation("Fiji")
my_fav_vacation("Fiji", "Winter")
my_fav_vacation(country = "Bali", season = "Fall")
```

```
def my_fav_holidayitems_list(holidaythings):
  if type(holidaythings) == list:
    for x in holidaythings:
      print("Some of my favorite holiday things are", x)
  else:
    print("Please pass a list object in this function")
```

```
holiday = ["Holiday Lights", "Snow", "Presents", "Break from School", "Hoy Cocoa", "Pumpkin Spice"]
holiday1 = 'Holiday Lights'
holiday2 = ("Holiday Lights", "Snow", "Presents", "Break from School", "Hoy Cocoa", "Pumpkin Spice")

my_fav_holidayitems_list(holiday)
my_fav_holidayitems_list(holiday1)
my_fav_holidayitems_list(holiday2)
```

```
#Return Values

def my_tip_calculator(subtotal, tax, pct = 18.5):
  total = subtotal + tax #Total with tax
  tip1 = subtotal * (pct/100) #Percentage of Subtotal
  tip2 = total * (pct/100) #Percentage of Total
  tip3 = 2 * tax # Double tax method

  cheaptip = min(tip1, tip2, tip3) #Get the minimum value out of all the tip methods

  print("For Total:", total, "Tip Method 1:", tip1, "Tip Method 2:", tip2, "Tip Method 3:", tip3)

  return(cheaptip)
```

```
my_tip_calculator(subtotal = 20.43, tax = 1.93, pct = 15)
```

```
tipamount = my_tip_calculator(20.43, 1.93)
```

```
print("My tip calculator said to tip: $", tipamount)
```
